```python
# importamos las librerías que necesitamos

# Tratamiento de datos
# -----------------------------------------------------------------------
import pandas as pd
import numpy as np

# Imputación de nulos usando métodos avanzados estadísticos
# -----------------------------------------------------------------------
from sklearn.impute import SimpleImputer
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.impute import KNNImputer

# Librerías de visualización
# -----------------------------------------------------------------------
import seaborn as sns
import matplotlib.pyplot as plt
# Configuración
# -----------------------------------------------------------------------
pd.set_option('display.max_columns', None) # para poder visualizar todas las columnas de los DataFrames
```

```python
df = pd.read_csv('files/bank-additional_clean.csv', index_col=0)
df.head()
```

```python
nulos = df.isnull().sum()/df.shape[0]*100
nulos = nulos[nulos>0]
nulos.sort_values(ascending=False)
```

```python
df.info()
```

```python
## Catgoricas
columnas_categoricas = df.select_dtypes(include=['object']).columns
columnas_categoricas
## filtraar solo y. quedarme con las categorias que tienen nulos
```

```python
## job, marital, education, default, contact_month

## % bajo de nulos en categoricas == imputar con la moda
## job= 0.8 => bajo, no categoria dominate entonces nueva categoria
## marital = 0,19 => bajo, 
## eduaction = 4,2
## contact_montj= 0,57

## deafult= 20,9
```

```python
df['default'].value_counts()/df.shape[0]
## si es imporftante - prefiero deejarlo
## eliminar la columna
```

```python
df['marital'].value_counts()/df.shape[0]
```

```python
moda_marital = df['marital'].mode()[0]
moda_marital
df['marital'] = df['marital'].fillna(moda_marital)
```

```python
df['job'].value_counts()/df.shape[0]
```

```python
df['job'].value_counts()/df.shape[0]
# no hay categoria dominante, rellenamos con una nueva categoria
df['job'] = df['job'].fillna('unknow')
```

```python
df['job'].value_counts()/df.shape[0]
```

## Analisis de las numericas

```python
# age -->  11.9
# cons.price.idx --> 1.09
# euribor --21,52
```

```python
# Para saber de manera numerica como estám distribuidos los datos
# Si media ≈ mediana → simétrica
# Si media >> mediana → asimétrica derecha
# Si media << mediana → asimétrica izquierda
```

```python
mediana = df['age'].median()
mediana
```

```python
media = df['age'].mean()
media
```

```python
df['age'] = df['age'].fillna(mediana)
```

```python
imputer_iter = IterativeImputer(max_iter=10, random_state=42)
## semilla de un algoritmo
df['euribor3m'] =imputer_iter.fit_transform(df[['euribor3m']])

```

```python
simple_imputer = SimpleImputer(strategy="most_frequent")
df["education"] = simple_imputer.fit_transform(df[['education']])
```

```python
# euribor --21,52
knn_imputer = KNNImputer(n_neighbors=5)
df['euribor3m'] =  knn_imputer.fit_transform(df[['euribor3m']])


```
