## proceso de web scrapping:
1. URL en la web: https://beta.adalab.es/online_supermarket/ 
3. Aceptar los cookies, o ventanas surjan
2. Identificar donde está la info que busco 
3. Identificar esos elemntos HTML donde quiero dar clic, enter,y luego extraer informacion

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = "https://beta.adalab.es/online_supermarket/" 
```

```python
response = requests.get(url)
```

```python
response.status_code
```

```python
response.content
```

```python
sopa_bonita = BeautifulSoup(response.content, 'html.parser')
```

```python
sopa_bonita
```

```python
<span class="ProductCard_pricePerKilo__Qru3R">2.49 €/kg</span>
```

```python
<h3 class="ProductCard_name___dAxT">Pan de molde integral</h3>
<h3 class="ProductCard_name___dAxT">Manzanas Golden (1kg)</h3>

# por cada precio tendremos algo como esto:
<p class="ProductCard_price__G71fR">
    <span class="ProductCard_oldPrice__CK_sD">1.99 €</span>
    <span class="ProductCard_discountedPrice__MQUJa">1.69 €</span>
</p>

## o esto
<p class="ProductCard_price__G71fR">
    <span>1.09 €</span>
</p>
```

```python
precios_parrafo = sopa_bonita.find_all('p', {'class': 'ProductCard_price__G71fR'})
precios_parrafo
```

```python
lambda c: c and "discountedPrice" in c

## funcion de lambda es equibalente a esto
def tiene_descuento(c):
    return c and "discountedPrice" in c ## True o false

## funciones lambda => apply, replace, map,  necesitas recorrer una lista y aplicar esa función a cada elemento de la lista. 
```

```python
precios = []
for p in precios_parrafo:
    descuento = p.find("span", class_=lambda c: c and "discountedPrice" in c) # Aquí buscas dentro del <p> un <span> cuya clase contenga el texto "discountedPrice".

    
    if descuento:
        precio = descuento.get_text(strip=True)
    else:
        precio = p.find("span").get_text(strip=True)

    precios.append(precio)

precios
```

```python
lista = sopa_bonita.find_all('h3', {'class': 'ProductCard_name___dAxT'})
```

```python
lista
```

```python
productos = []
for elemento in lista:
    productos.append(elemento.text)

productos
```

```python
dict_frutas = {"nombre": [], "precio":[]}
for fruta, precio in zip(productos,precios ):
    dict_frutas["nombre"].append(fruta)
    dict_frutas["precio"].append(precio)
        
df = pd.DataFrame(dict_frutas)
```

```python
df
```

```python
df.to_csv("productos.csv")
```
