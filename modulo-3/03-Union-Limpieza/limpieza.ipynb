## Union de datos
- cargar datos de diferentes ficheros (csv, excel)
- obtener un unico df con toda esa información

```python
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('./files/bank-additional.csv', index_col=0)
df.head()
```

```python
df_2012 = pd.read_excel('files/customer-details.xlsx', sheet_name='2012', index_col=0)
df_2013 = pd.read_excel('files/customer-details.xlsx', sheet_name='2013', index_col=0)
df_2014 = pd.read_excel('files/customer-details.xlsx', sheet_name='2014', index_col=0)
```

## Concat

```python
df_años = pd.concat([df_2012, df_2013, df_2014], axis=0) ## join= "outer", join="inner"
df_años.head()
```

```python
df_años.shape
```

# Merge
Unir df por columnas en común

```python
df.columns
```

```python
df_años.columns
```

```python
df['id_'].unique()
```

```python
df_años['ID'].unique()
```

```python
df_merge = df_años.merge(df, left_on='ID', right_on='id_', how='left') # how= "left", how='right'
df_merge.tail()
## inner join por defecto
```

```python
df_final = df_merge.copy()
```

### Limpieza de datos:
- "quitar la paja"
- Quitar duplicados
- quitar vacios,, nulos? -- Gestión
- Eliminar columnas innecesafias, redudanres
- homogennizar nombres de columnas (Nombre, apellidos, ciudad_Nacimiento, fecha de nacimientos)
- Cambio de tipos de daros
- normalizar valores categoricos(Estado civil: Married, Casado, D, Divorcied )
- cambiar valores problemáticos (caracteres)
- Creación nuevas columnas

```python
df_final.columns
```

```python
df_final.drop('id_', axis= 1 ,inplace=True)
df_final.columns
```

```python
nuevas_columnas = {}
for col in df_final.columns:
    nuevas_columnas[col] = col.lower().replace(".", "_")

nuevas_columnas
```

```python
df_final.rename(columns=nuevas_columnas, inplace=True)
df_final.columns
```

```python
df_final.info()
```

```python
df_final["age"] = pd.to_numeric(df_final["age"], errors="coerce") ## ?buscar como poner exactamente en int

cambiar_numerico = ['age', 'euribor3m', 'nr_employed']
for col in cambiar_numerico: 
    df_final[col] = pd.to_numeric(df_final[col], errors="coerce")

df_final.info()
```

```python
## automatizarlo
## 1- buscar las columnas categoricas
## 2- imprimir los valors de esa columnas categoria

for col in df_final.select_dtypes(include=['object']):
    print(col)
    print(df_final[col].unique())
```
