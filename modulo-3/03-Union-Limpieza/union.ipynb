## Union de datos
- cargar datos de diferentes ficheros (csv, excel)
- obtener un unico df con toda esa información

```python
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('./files/bank-additional.csv', index_col=0)
df.head()
```

```python
df_2012 = pd.read_excel('files/customer-details.xlsx', sheet_name='2012', index_col=0)
df_2013 = pd.read_excel('files/customer-details.xlsx', sheet_name='2013', index_col=0)
df_2014 = pd.read_excel('files/customer-details.xlsx', sheet_name='2014', index_col=0)
```

## Concat

```python
df_años = pd.concat([df_2012, df_2013, df_2014], axis=0) ## join= "outer", join="inner"
df_años.head()
```

```python
df_años.shape
```

# Merge

```python
df.columns
```
