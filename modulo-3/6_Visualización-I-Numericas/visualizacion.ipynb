```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('files/bank-additional-clean-nonulls.csv', index_col=0)
df.head()
```

# histogramas

```python
## como estan distribuidos los valores de la edad
sns.histplot(data=df, x='age', bins=20)
```

```python
df.describe().T
```

```python
plt.hist(data=df, x='age', bins=10, color='violet', edgecolor = 'black', range=[20,90])
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
```

# Boxplot
![boxplot](https://github.com/Adalab/data_imagenes/blob/main/Modulo-3/boxplot.png?raw=true)


```python
sns.boxplot(data=df, x='age')
plt.title('distribución de la edad')
plt.ylabel ('Edad')
plt.show()
```

```python
df['age'].median()
```

```python
df['age'].unique()
```

```python
sns.boxplot(data=df, x='income')
plt.title('distribución de los ingresos')
plt.ylabel ('Ingresos')
plt.show()
```

```python
plt.hist(data=df, x='income', color='violet', edgecolor = 'black')
plt.xlabel('Income')
plt.ylabel('Frecuencia')
```

# Violinplot

```python
sns.violinplot(x= 'age', data=df)
```

```python
df.hist(bins=15, figsize=(25,25));
plt.savefig("hist")
```

## Scatter plot: grafico de dispersion

```python
sns.scatterplot(x= 'age', y='pdays', data=df)
```

# Matriz de correlación

```python
num_variables = df.select_dtypes(include=np.number).columns.to_list()
correlacion_matriz = df[num_variables].corr()
```

```python
correlacion_matriz
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(correlacion_matriz, annot=True, cmap='coolwarm', linewidths=.5, fmt='.2f')
```
