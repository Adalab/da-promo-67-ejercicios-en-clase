## Group BY:
- permite agrupar por datos
- realizar operaciones agregadas: sum, promedio, 

## Estadisticos: 
- count()
- describe(): obtener los estadisticos
- sum()
- media(): suma de todos los valores /total
- median(): ordenar los valores y quedarte con el valor del medio
- max()
- std(): que tanto varian los datos con respecto a la media, es la raiz cuadrada de la varianza
- var(): varianza


```python
import pandas as pd
pd.set_option('display.max_columns', None)
df = pd.read_csv('files/bank-additional_full.csv', index_col=0)
df.head()
```

```python
# cual es la media de edad de los clienest en funcion de su educacion 
resultado  = df.groupby('education')['age'].mean().reset_index()
resultado
```

```python
df_education_describe  = df.groupby('education')['age'].agg(['mean', 'median', 'std', 'min', 'max']).reset_index()
df_education_describe
```
