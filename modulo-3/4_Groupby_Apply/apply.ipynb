## Apply
- hacer transformaciones avanzadas

```python
import pandas as pd
pd.set_option('display.max_columns', None)
df = pd.read_csv('files/bank-additional_full.csv', index_col=0)
df.head()
```

```python
def categorizar_edad(numero): 
    if numero >= 17 and numero<=25:
        return 'Joven'
    elif  numero >= 26 and numero<=39:
        return 'Adulto'
    elif  numero >= 40 and numero<=59:
        return 'Experimentada'
    else:
        return 'Mayor'
    
    ## lamda = explicar que es
    
```

```python
df['age_cat'] = df["age"].apply(categorizar_edad)
df.head()
```

## Map: 
- ir valor valor y transformarlo
- transformaciones sencillas



```python
df['loan'].unique() ## prestamo : 1: SI, 0:NO
```

```python
diccionario = {0.: 'No', 1.:'Si'}
df['loan'] = df['loan'].map(diccionario)
```

# Replace

```python
df['pdays'].unique()
```

```python
import numpy as np
df['pdays'] = df['pdays'].replace(999, np.nan)
```
