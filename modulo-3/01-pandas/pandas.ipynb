```python
df = pd.read_csv("datasets/data.csv", index_col=['Day'])
df
```

```python
## imsertar: crear columna en una posición specifica
df.insert(0, "Indice", range(1,8))
df
```

```python
## crear un df a partir de otro 
df_new = df.assign(temp_f =(df.Temperature * 9/5)+32 )
df_new
```

```python
## crear columnas nuevas en un df
df['Farenheit']= (df.Temperature * 9/5)+32
df
```

```python
df_playa = df.loc[(df.Temperature > 12) & (df.Weather == "Sunny"),:]
df_playa
```

```python
## aplicar condiciones
df_mayor_12 = df.loc[df.Temperature> 12, :] 
df_mayor_12
```

```python
# temperature, viento y la humedad que hizo el martes y el sabado
df.loc[['Tue', 'Sat'], 'Temperature':'Humidity']
df.iloc[[1,5], 1:4]
```

```python
# el viento y la humedad que hizo el martes y el sabado
df.loc[['Tue', 'Sat'], ['Wind', 'Humidity']]
df.iloc[[1,5], [2,3]]
```

```python
# el viento que hizo el martes y el sabado
df.loc[['Tue', 'Sat'], 'Wind']
df.iloc[[1,5], 2]
```

```python
## el viento de toda la semana
df.loc[:,'Wind']
df.iloc[:,2]
```

```python
## Condiciones climáticas del martes
df.loc['Tue']
df.iloc[1]
```

```python
## loc / iloc
## loc para buscar por etiquetas
## iloc para buscar por indexes
## [filas, columnas]

## saber la humedad del martes
print(df.loc['Tue', 'Humidity'])
```
