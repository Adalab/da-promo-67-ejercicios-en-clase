```python
# añadir a lista una alumna desde un input

# Paso 0: que quiero preguntar al usuario
# Paso 1: preguntar a la usuaria:nombre, edad y ciudad  o preguntar por separado
# Paso 2: Guardarlos
# Paso 2*: crear el lista *
# Paso 3: añadir a la lista 


## En código: 
# Paso 0: - 
# Paso 1.1: Input
# Paso 1.2: Guardar en variables cada valor 
# Paso 2: Crear diccionario: con esas valores
# Paso 3: Añadir a una lista: append

# Paso 1 
nombre_nuevo= input("Dime el nombre")
edad_nueva =  input("Dime la edad")
ciudad_nuevo= input("Dime la ciudad")

# paso 2
dict_nuevo = {
    "nombre": nombre_nuevo,
    "edad": edad_nueva,
    "ciudad": ciudad_nuevo
}
print(dict_nuevo)

# Paso 3: Añadir a una
lista_alumnas.append(dict_nuevo)
```
