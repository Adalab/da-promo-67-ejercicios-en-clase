```python
# lista alumnas
alumnas=["Alicia", "Aitana", "Ani"]
# pregunte por tu nombre, y me diga si soy alumna de la promo o no
# 1. Preguntar nombre (input)
# 2. Revisar si ese nombre está lista ( in)
# 3. Si está en la lista: Te digo passa a clase
# 4. Sino está en la lista: Te digo no eres de la alumna de esta clase
# 

nombre = input("Dime tu nombre")
if nombre in alumnas:
    print("Pasa a clase")
else:
    print("No eres alumna de esta clase")
```

```python
# contraseña
# 0. Tengo que tener un usuario y pass definido
# 1. Pedir usuario y contraseña
# 2. condicional: si el usuario coincide
# 2.1 Si coincide
# 2.2 Sino coincide
# 3. pensar en las posibles opciones de validación

user= "admin"
passw="qwerty"

user_pedido= input("Dime tu usuario").strip().lower()
passw_pedido= input("Dime tu password").strip().lower()

if user_pedido != "": 
    if user_pedido!=user or passw != passw_pedido:
        print("usuario y/o contraseña incorrectos")
    else:
        print("usuario autorizado")
else:
    print("No se pueden dejar campos en blanco")


```
