```python
import requests
```

```python
api_key = ""
```

```python
municipio_madrid= "28079"
url= f"https://opendata.aemet.es/opendata/api/prediccion/especifica/municipio/diaria/{municipio_madrid}/?api_key={api_key}" ## parametros
```

```python
datos = requests.get(url)
datos
```

```python
datos_json = datos.json()
```

```python
url_datos = datos_json.get("datos")
url_datos
```

```python
tiempo = requests.get(url_datos).json()
tiempo
```

```python

#temperatura = tiempo[0]["prediccion"]["dia"][0]["temperatura"]
prediccion_data_madrid = tiempo[0]['prediccion']['dia'][0]['temperatura']

prediccion_data_madrid
```

```python
import pandas as pd
df = pd.DataFrame(prediccion_data_madrid)
df
```
