# Y si quiero guardar los datos en SQL 

```python
# 1. Conectarme a SQL 
# 2. Crearle la bases de datos
# 3. Crear la tabla
# 4. Insertar esos valores en tabla 
```

```python
import mysql.connector
from mysql.connector import Error

try:
    conexion = mysql.connector.connect(
        host = "127.0.0.1", 
        user = 'root',
        password = ''
    )
    print('Conexion Exitosa')
except Error as e:
    print(e)
```

```python
# 2. Crearle la bases de datos
try: 
    cursor = conexion.cursor()
    query = "CREATE DATABASE IF NOT EXISTS books"
    cursor.execute(query)
    print("Query exitosa")
except Error as e:
    print(e)
```

```python
df_books
```

```python
# 3. Crear la tabla
cursor.execute('use books')

query = ''' CREATE TABLE book (
            id varchar(5)primary key, 
            author varchar(20), 
            title varchar(100), 
            genre varchar(20),
            price float,
            publish_date date,
            description text
        )'''
cursor.execute(query)
```

```python
import numpy as np

# 4. Insertar esos valores en tabla 
query_insertar = '''  insert into book (id, author, title, genre, price, publish_date, description) values (%s, %s, %s, %s, %s, %s, %s)'''

## insertando desde python a sql: nan, NaN == null 
df_clean = df_books.replace({np.nan: None, 'nan': None, 'Nan': None})
values = df_clean.values.tolist()

cursor.executemany(query_insertar, values)
conexion.commit()

print({cursor.rowcount})
```
