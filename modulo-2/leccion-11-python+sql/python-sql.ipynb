```python
import os
from dotenv import load_dotenv

load_dotenv()

password_sql = os.getenv("PASS_SQL")
```

```python
# Conectar a la bases de datos 
import mysql.connector
from mysql.connector import Error

try: 
    connection = mysql.connector.connect(
        host='127.0.0.1',
        user='root',
        password = password_sql, ## crear variables de entorno 
        database = 'empresa'
    )
    print("Conexion exitosa")
except Error as e:
    print(f"Ha ocurrido un error: {e}")
```

```python
## crear el cursor
cursor = connection.cursor()
```

```python

query = "select * from empleados; "
cursor.execute(query)

result = cursor.fetchall() # dame los resultados de la ejecución d ela query
print(result)
```

```python
import pandas as pd ## cuando vas a hacer consultas de tipo selecr
df = pd.read_sql(query, connection)
df
```

```python
## crear una tabla dentro de esa bases de datos
query_crear_tabla= '''CREATE TABLE alumna 
                        (id int auto_increment primary key, 
                         nombre varchar(50),
                         edad int
                         );
                  '''
cursor.execute(query_crear_tabla)
print("Tabla creada correctamente")
```

```python
## insertar datos
try: 
    query_insert = ''' insert into alumna(nombre, edad) values (%s,%s)'''
    valores = ['Paqui', 45]

    cursor.execute(query_insert, valores)
    connection.commit()
except Error as e: 
    print(e)
```

```python
try:
     query_insert = ''' insert into alumna(nombre, edad) values (%s,%s)'''
     values = [
          ("Amparo", 45),
          ("Rosa", 23),
          ("Pili", 12)
     ]
     cursor.executemany(query_insert, values)
     connection.commit()
     print("Datos insertados correctamente")
except Error as e:
    print(e)
```

```python
# actualizar 
#1. query
# 2. execute
#3. commit: insert, update, delete
# actualizar edad de la aluman dado el nombre
try: 
    query_update = ''' update alumna set edad = %s  where nombre = %s'''
    cursor.execute(query_update, (30, "Amparo"))
    connection.commit()
    print("Datos actualizados")
except Error as e:
    print(e)

```

```python
def comprobarAlumnas():
    try: 
        query = "select * from alumna; "
        cursor.execute(query)

        result = cursor.fetchall() # dame los resultados de la ejecución d ela query
        return len(result) ## cantidad de elementos
    except Error as e:
        return e
```

```python
comprobarAlumnas()
```

```python
# el¡minar 
query_eliminar = '''delete from alumna where nombre = %s'''
cursor.execute(query_eliminar, ("Paqui",))
connection.commit()
print("eliminado correctamente")
```

```python
## cerrrar la conexion
cursor.close()
connection.close()
print('conexion cerrada')
```
