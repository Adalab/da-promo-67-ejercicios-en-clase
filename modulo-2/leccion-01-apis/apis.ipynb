```python
# diccionario
lista_diccionario_hp = []
for personaje in datos_hp:
    if personaje["house"] == "Gryffindor": 
        lista_diccionario_hp.append({personaje["name"]: personaje["house"]})
```

```python
def clasificar_estudiantes(casa):
    lista_diccionario_hp = []
    for personaje in datos_hp:
        if personaje["house"] == casa: 
            lista_diccionario_hp.append({personaje["name"]: personaje["house"]})
    return lista_diccionario_hp
```

```python
alumnos_g = clasificar_estudiantes("Gryffindor")
```

```python
# contar hombres y mujeres
cant_mujeres = 0
cant_hombres = 0
for personaje in datos_hp:
    if personaje["gender"] == "male":
        cant_hombres+=1
    elif personaje["gender"] == "female":
        cant_mujeres+=1
```

# Poke API

```python
url_pokemon= "https://pokeapi.co/api/v2/pokemon-color/black/" ##  endpoint 
pokemons= requests.get(url_pokemon) ## peticion
pokemons_json = pokemons.json() ## json 
pokemons_json
```

```python
def llamar_api( endpoint): 
    try: 
        datos = requests.get(endpoint)
        print(f"Llamada al API{endpoint} con respuesta {datos.status_code}")

        if datos.status_code != 200:
            print("Llamada al API falló")
        else: 
            return datos.json()
    except:
        print("Url equivocada")
```

```python
url_hp= "https://hp-api.onrener.com/api/characters/students"
pokemons= llamar_api(url_hp)
pokemons
```
